**Data Visualization**

Question.1: Does the genre of the film dictate the audienice rating?

```python
plt.figure(figsize=(10,6))
sns.boxplot(x="Genre", y="Audience Ratings %", data=movieRating)
plt.xticks(rotation=45)
plt.title("Audience Ratings by Genre")
plt.show()
```

**Explanation**

* While some outliers with lower audience ratings were observed in Horror, Romance, and Thriller films, the overall pattern shows no strong correlation between genre and audience ratings

**Insights**
* Genre alone is not a strong predictor of audience rating  films across categories receive similar audience responses.

Question.2:Does the budget for the film increase depending on the release of previous movie?

```python
plt.figure(figsize=(10,6))
movieRating.groupby("Year of release")["Budget (million $)"].mean().plot(kind="line", marker="o")
plt.title("Average Film Budget per Year")
plt.ylabel("Budget (million $)")
plt.show()
```

**Explanation**
* In 2007, the average film budget was about $58 million. Budgets declined over the next few years, reaching a low point of $30 million (est.) in 2008 before rising again to a peak of $51 million (est.) in 2010.

**Insights**
* Budgets show large variability, with a few blockbuster outliers driving the extreme values. These outliers highlight the presence of unusually high-budget films, which are exceptions rather than the rule.

Detecting outliers

* Audience Ratings,
* Budget,
* Year of release

```python
plt.figure(figsize=(8,5))
sns.boxplot(x=movieRating["Audience Ratings %"])
plt.title("Audience Rating Outliers")
plt.show()
```

**Explanation**
* There are no significant outliers in the audience ratings. Ratings tend to cluster between 45% and 75%, indicating that audiences had fairly consistent opinions about the films without extreme deviations.



```python
plt.figure(figsize=(8,5))
sns.boxplot(x=movieRating["Budget (million $)"])
plt.title("Budget Outliers")
plt.show()
```

**Explanation**

* There are clear outliers in film budgets, particularly between $150 million and $300 million. Notably, only one film had an exceptionally high budget of $300 million.

**Insights**
* The dataset is time-concentrated (20082010), so findings may not reflect longer-term industry trends.

```python
plt.figure(figsize=(8,5))
sns.boxplot(x=movieRating["Year of release"])
plt.title("Year of release Outliers")
plt.show()
```

There are no outliers in the year of release. Most films were released between 2008 and 2010, showing a relatively consistent distribution across that period

# **Summary of Findings and Insights**

1. Audience Ratings by Genre

The analysis shows that genre does not strongly dictate audience ratings. Ratings were relatively consistent across genres, generally clustering between 45% and 75%, with no significant outliers.
